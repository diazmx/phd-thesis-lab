## 18-CN&S - Descomposición en k-núcleos.

```python
### Importar bibliotecas

import igraph as ig
```

```python
### Cargar grafo

file_name = "Karate_club_from_gephi.graphml"
g = ig.read(file_name)
print(g.summary())
```

```python
### Obtener la descomposición en k-núcleos

"""
def k_core(graph, *args)

devuelve los k-núcleos del grafo.

El método acepta un número arbitrario de argumentos que representan los índices
deseados de los k-núcleos que se devolverán. Los argumentos también pueden ser
listas o tuplas.

Retorna un único objeto 'Graph' si solo se proporcionó un argumento 
entero; de lo contrario, el resultado es una lista de objetos 'Graph' que
representan los k-cores deseados en el orden en que se especificaron los argumentos.

Si no se proporciona ningún argumento, devuelve todos los k-núcleos en orden
creciente de k.
"""

g.k_core()
```

```python
for kcore in g.k_core():
    print(kcore.summary())
```

```python
# Cálcula los cores del grafo
kcores = g.k_core()

print("Tamaño de la lista:", len(kcores))
print("Maximo grado:", g.vcount())
```

```python
print(g.summary())
ig.plot(g, bbox=(400,400))
```

```python
k = 3
kcores = g.k_core(k)

print(type(kcores))
print(kcores.summary())
ig.plot(kcores, bbox=(400,400))
```

```python
# Cálcula los cores del grafo
k_list = range(6)
kcores = g.k_core(k_list)

print("Tamaño de la lista:", len(kcores))
print("Maximo grado:", g.vcount())

for kcore in kcores:
    print(kcore.summary())
```

```python
### Encontrar el coreness de cada nodo (k-shell)

print(len(g.coreness()), g.coreness())
```

```python
### Agregar coreness como atributo de nodos

g.vs["kshell"] = g.coreness()
print(g.vs["kshell"])
```

```python
### Generar k-shell

nodes_in_shell = g.vs.select(kshell=2) # Selección de nodos con coreness = 1

shell_1_subgraph = g.subgraph(vertices=nodes_in_shell)
print(shell_1_subgraph.summary())
ig.plot(shell_1_subgraph, bbox=(300,300))
```

```python
### Generar k-shell

nodes_in_shell = g.vs.select(kshell=4) # Selección de nodos con coreness = 1

shell_4_subgraph = g.subgraph(vertices=nodes_in_shell)
print(shell_4_subgraph.summary())
ig.plot(shell_4_subgraph, bbox=(300,300))
```

```python
### Visualización

layout_ = g.layout_lgl()

color_dict = {1: "blue", 2: "orange", 3:"red", 4: "yellow"}
visual_style = {}
visual_style["vertex_size"] = 20
visual_style["vertex_color"] = [color_dict[clase] for clase in g.vs["kshell"]]
visual_style["vertex_label"] = g.vs["label"]
visual_style["layout"] = layout_
visual_style["bbox"] = (300, 300)
visual_style["margin"] = 20
ig.plot(g, **visual_style)
```
