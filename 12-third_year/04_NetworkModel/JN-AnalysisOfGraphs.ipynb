```python
### Graph Analysis
FILENAME = ORIGIN_PATH+"12-third_year/00-Data/02-HC/02-Graphs/binet-HC-Rw.gml"
g = ig.read(FILENAME)
print("Is bipartite? ", g.is_bipartite())

print("|V| =", g.vcount())
print("|E| =", g.ecount())

print()
print("Nodes")
conta = 0
for node in g.vs():
    if conta < 10:
        print(node)  
        conta += 1
    else:
        break

print()
print("EDGES")
conta = 0
for edge in g.es():
    if conta < 10:
        print(edge)  
        conta += 1
    else:
        break
```
