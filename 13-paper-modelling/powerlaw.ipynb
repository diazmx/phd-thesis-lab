```python
import powerlaw
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
# Crear red de ejemplo
#G = nx.barabasi_albert_graph(1000, 3)
df = pd.read_csv("git-grafo.csv")
G = nx.from_pandas_edgelist(df)
degrees = [d for n, d in G.degree()]

# Ajustar la distribucin
fit = powerlaw.Fit(degrees, discrete=True)

# Estimar alpha y xmin
alpha = fit.alpha
xmin = fit.xmin

print(f"Alpha: {alpha:.3f}")
print(f"xmin: {xmin}")

# Visualizar el ajuste
fig = fit.plot_pdf(color='b', label='Datos')
fit.power_law.plot_pdf(color='r', linestyle='--', ax=fig, label='Ley de potencias ajustada')
plt.legend()
plt.xlabel('Grado')
plt.ylabel('P(D >= k)')
plt.title('Ajuste de ley de potencias a la distribucin de grados')
plt.show()


```
